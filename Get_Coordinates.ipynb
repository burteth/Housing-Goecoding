```python
def clean_data(addy):
    x = addy.lower()
    if "near" in x:
        head, sep, tail = x.partition('near')
        x = head
    if "bath" in x:
        head, sep, tail = x.partition(' - ')
        x = tail
    if "unit" in x:
        head, sep, tail = x.partition('unit')
        x = head
    if "apt" in x:
        head, sep, tail = x.partition('apt')
        x = head
    if "#" in x:
        head, sep, tail = x.partition('#')
        x = head
    return(x)

```

```python
a["updated_addresses"] = a["address"].apply(clean_data)
```
